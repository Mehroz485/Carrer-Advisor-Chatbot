```python
import ollama
 
SYSTEM_PROMPT = """
You are CareerBot, a professional Career Advisor AI assistant.
Only answer questions related to careers, jobs, professional
development, resumes, interviews, salary negotiation,
networking, and career transitions.
If asked about anything outside these topics, respond with:
'That is outside my expertise as a Career Advisor...'
"""
 
conversation_history = []
 
def chat(user_message):
    conversation_history.append({"role": "user", "content": user_message})
    response = ollama.chat(
        model="phi3",
        messages=[{"role": "system", "content": SYSTEM_PROMPT},
               *conversation_history])
    assistant_reply = response["message"]["content"]
    conversation_history.append({"role": "assistant", "content": assistant_reply})
    return assistant_reply
 
print("CareerBot - AI Career Advisor (type 'exit' to quit)")
 
while True:
    user_input = input("You: ").strip()
    if not user_input: continue
    if user_input.lower() == "exit": break
    print(f"CareerBot: {chat(user_input)}")





```
